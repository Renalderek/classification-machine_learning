```python
import pandas as pd

# membaca dataset menjadi dataframe

data = pd.read_csv('Salary_Data.csv')

import numpy as np

X = data['YearsExperience']
y = data['Salary']

# Mengubah Bentuk atribut
X = X[:,np.newaxis]

from sklearn.svm import SVR

# Membangun Model dengan paramater C,gamma,dan kernel
model = SVR(C=1000,gamma=0.5,kernel='rbf')

# melatih model dengan fungsi fit
model.fit(X,y)

import matplotlib.pyplot as plt

# matplotlib untuk memvisualisasikan model yang dibuat
plt.scatter(X,y)
plt.plot(X,model.predict(X))
```
